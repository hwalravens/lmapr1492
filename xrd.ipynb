```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("g23npTJNAhDPnQMr") as m:

    # Structure and doc for material id
    struct = m.get_structure_by_material_id("mp-9363")
XRD = XRDCalculator(wavelength='CuKa')
```

# 1.Angles de diffraction et intensités

```python
pattern = XRD.get_pattern(struct,scaled=True)
print(pattern)
```

Le premier pic de diffraction correspond à un angle 2θ de 11.4721946 degrés et une intensité de 21.7517160. Le deuxième pic à un angle 2θ de 23.06143487 degrés et une intensité de 21.0210048. Le troisème pic à un angle 2θ de 24.61514654 degrés et une intensité de 4.22546635. Les intensités sont mises à l'échelle, le pic maximum est fixé à une valeur de 100.

# 2. Indices de Laue (h,k,l)

```python
plot = XRD.get_plot(struct)
print(plot)
```

Le premier pic de diffraction a un indice de Laue (1,1,1), le deuxième pic de diffraction a un indice de Laue (2,2,2) et le troisième pic de diffraction a un indice de Laue (1,0,0).
